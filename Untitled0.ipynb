Caricamento dati e descrizione dataset

```
import pandas as pd

df = pd.read_csv('ML-CUP19-TR.csv', skiprows=7, header=None)
n_records = len(df)

df.describe()
```

Definizione variabili di input e di output e normalizzazione output con minimo e massimo

```
import numpy as np
from sklearn.preprocessing import MinMaxScaler



X = np.array(df.iloc[:,1:-2])
y = np.array(df.iloc[:,-2:])

scaler = MinMaxScaler()
y = scaler.fit_transform(y)
```

Prova regressione con rete neurale

```
from NeuralNetwork import MultiLayerPerceptron
import time

NN = MultiLayerPerceptron(n_hidden_nodes=50,
                          hid_activ_fun='sigmoid',
                          learning_rate=10**-1,
                          momentum=0.7,
                          regularization=10**-7,
                          evaluate=True,
                          verbose=True,
                          epochs=100)

start = time.time()
NN.fit(X,y)
end = time.time()
print('Time for fitting {} minutes {} seconds'.format(int((end-start)/60), int(end-start)%60))

y_pred = NN.predict(X)
```

Plot di gradiente e MSE durante la fase di training

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

start_epoch = 100

plt.subplot(1,2,1)
plt.plot(NN.MSE_values_train[start_epoch:])
plt.xlabel('Epochs')
plt.title('MSE')

plt.subplot(1,2,2)
plt.plot(NN.gradient_values[start_epoch:])
plt.xlabel('Epochs')
plt.title('Gradient')

plt.show()
```

Distribuzione dei residui

```
from seaborn import distplot
from statistics import mean, stdev

#y = scaler.inverse_transform(y)
#y_pred = scaler.inverse_transform(y_pred)

residuals = list()

for i in range(n_records):
    residuals.append(np.linalg.norm(y[i]-y_pred[i])/np.linalg.norm(y[i]))
    
print('Residuals')
print('Mean: {0:.4f}, Standard deviation: {1:.4f}, Min: {2:.4f}, Max: {3:.4f}'
      .format(mean(residuals), stdev(residuals), min(residuals), max(residuals)))

plt.figure(figsize=(8,8))
distplot(residuals)
plt.title('Residuals distribution')

plt.show()
```
